```python
def round_dataset(df, column_name, decimals):
    """
    (1) Takes in a dataset and rounds every value to 3.
    (2) Formats one specific column to x number of decimals given by user

    Args:
        df (dataframe): dataframe to be formatted
        column_name (string): Name of the column to be formatted with x number of decimals
        decimals (_type_): _description_

    Returns:
        dataframe: returns formatted dataframe
    """
    df = round(df, 3)
    df[column_name] = df[column_name].apply(lambda x: round(x, decimals))
    print("✅ Dataset formatted..")
    return df
```
