```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import swat
from time import time



# Importing preprocessed dataset
import sys
sys.path.append("..")
from Functions.UNSW_DF import *
train, test = DF_original()
```

```python
num_features = train.columns[train.dtypes != 'object']
fig, axes = plt.subplots(1, 3, figsize=(50, 40), sharey=True, dpi=100)
for i in num_features:
    plt.figure()
    sns.displot(train[i])
```

```python
# Correlation plot
corr = train.corr()
sns.set_theme('notebook')

mask = np.triu(np.ones_like(corr, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

plt.figure(figsize=(20,12))
sns.heatmap(corr, mask=mask, cmap=cmap, center=0, square=True, linewidths=1)
plt.savefig('Dataset Correlation.png', dpi=300)
plt.show()
```
