```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../ANN/ANN_200E_3L(50-25-1)_AF(Relu-Sigmoid).csv")
dataset
```

```python
sns.set(rc={'figure.figsize':(18, 10)})
```

```python
barplot = sns.barplot(x="accuracy_train", y="epochs", data=dataset)
```

```python
sns.set_style("darkgrid") # or whitegrid
sns.lineplot(data = dataset, x = "accuracy_train", y = "epochs")
```

```python
pivot = dataset.pivot_table(
    index=["epochs"],
    columns=["accuracy_train"],
    values="no_layers",
    aggfunc=np.average)
sns.heatmap(pivot)
```

```python
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.scatterplot(data=dataset, x='accuracy_train', y='epochs',   hue='no_layers', palette='Set2', s=60)
#ax.set(xlabel='Height (cm)', ylabel='Weight (kg)')
#ax.legend(title='Country', title_fontsize = 12) 
plt.show()
```

```python
# library & dataset
import seaborn as sns

# use the function regplot to make a scatterplot
sns.regplot(x=dataset["epochs"], y=dataset["accuracy_train"])
```

```python
groups = dataset.groupby("epochs")
for name, group in groups:
    plt.plot(group["epochs"], 
             group["accuracy_train"], 
             group["accuracy_test"], 
             #group["precision_train"], 
             #group["precision_test"],
             #group["F1_train"],
             #group["F1_test"],
             #group["recall_train"],
             #group["recall_test"],
             #group["runtime(s)"],
             marker="o", 
             linestyle="") # Legend: label = name
plt.legend()
```
