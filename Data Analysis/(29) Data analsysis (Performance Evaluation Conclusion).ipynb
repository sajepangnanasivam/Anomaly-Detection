```python

import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_org = pd.read_csv("PE_conclusion.csv")
```

```python
df_metrics = df_org.drop(["Parameters", "runtime(s)", "runtime(s)_GS"], axis=1)
df_metrics_melt = pd.melt(df_metrics, id_vars="Algorithm", var_name="metric", value_name="score")
```

## Barplot

```python
ax_line = sns.set(rc={"figure.figsize":(24, 12), 'axes.facecolor':'white',})
ax_line = sns.set_style("whitegrid")
ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = df_metrics_melt, palette="viridis")
for container in ax_line.containers:
    ax_line.bar_label(container)


    
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0)
ax_line.set_xlabel("Algorithm",fontsize=20)
ax_line.set_ylabel("Score", fontsize=20)

ax_line.figure.savefig("Analysis_Conclusion/Comparison_Barplot.png", dpi=400)
```

```python
def show_values(axs, orient="v", space=.5):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.0f}'.format(p.get_height())
                ax.text(_x, _y, value, ha="center") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.1f}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```
