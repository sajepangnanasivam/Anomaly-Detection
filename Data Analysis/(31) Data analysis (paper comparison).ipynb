```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("..")

from Functions.UNSW_DF import * 
```

```python
df_org = pd.read_csv("Analysis_paper/paper_comp.csv")
df_org
```

```python
ANN = df_org[0:2]
SVM = df_org[2:4]
KNN = df_org[4:6]
LR = df_org[6:8]
DT = df_org[8:10]
RF = df_org[10:12]
ANN_melt = pd.melt(ANN, id_vars="Algorithm", var_name="metric", value_name="score")
SVM_melt = pd.melt(SVM, id_vars="Algorithm", var_name="metric", value_name="score")
KNN_melt = pd.melt(KNN, id_vars="Algorithm", var_name="metric", value_name="score")
LR_melt = pd.melt(LR, id_vars="Algorithm", var_name="metric", value_name="score")
DT_melt = pd.melt(DT, id_vars="Algorithm", var_name="metric", value_name="score")
RF_melt = pd.melt(RF, id_vars="Algorithm", var_name="metric", value_name="score")

#df_metrics = df_org.drop(["Parameters", "runtime(s)", "runtime(s)_GS"], axis=1)
df_metrics_melt = pd.melt(df_org, id_vars="Algorithm", var_name="metric", value_name="score")
```

```python
ANN_melt.dropna(subset = ["score"], inplace=True)
SVM_melt.dropna(subset = ["score"], inplace=True)
KNN_melt.dropna(subset = ["score"], inplace=True)
LR_melt.dropna(subset = ["score"], inplace=True)
DT_melt.dropna(subset = ["score"], inplace=True)
RF_melt.dropna(subset = ["score"], inplace=True)
```

```python
def legend_and_labels_fix(xlabel=""):
    for container in ax_line.containers:
        ax_line.bar_label(container, padding=1)
    ax_line.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
    ax_line.set_xlabel(xlabel, fontsize=16)
```

```python
ax_comb = sns.set(rc={"figure.figsize":(34, 20), 'axes.facecolor':'white',})
ax_hist = sns.set_style("whitegrid")

fig, axs = plt.subplots(ncols=2, nrows=3)

ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = ANN_melt, palette="viridis", ax=axs[0, 0])
legend_and_labels_fix(xlabel="Artificial Neural Network")

ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = SVM_melt, palette="viridis", ax=axs[0, 1])
legend_and_labels_fix(xlabel="Support Vector Machines")
    
ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = KNN_melt, palette="viridis", ax=axs[1, 0])
legend_and_labels_fix(xlabel="K-Nearest Neighbor")
    
ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = LR_melt, palette="viridis", ax=axs[1, 1])
legend_and_labels_fix(xlabel="Logistic Regression")
    
ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = DT_melt, palette="viridis", ax=axs[2, 0])
legend_and_labels_fix(xlabel="Decision Trees")
    
ax_line = sns.barplot(x = "Algorithm", y = "score", hue = "metric", data = RF_melt, palette="viridis", ax=axs[2, 1])
legend_and_labels_fix(xlabel="Random Forest")


    
#plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0)
#ax_line.set_xlabel("Algorithm",fontsize=20)
#ax_line.set_ylabel("Score", fontsize=20)

ax_line.figure.savefig("Analysis_paper/Paper_comparison.png", bbox_inches='tight', dpi=400)
```

```python
ax_comb2 = sns.set(rc={"figure.figsize":(50, 24), 'axes.facecolor':'white',})
ax_comb2 = sns.set_style("whitegrid")

dataset = df_metrics_melt

fig, axs = plt.subplots(ncols=4, nrows=2)
ax_comb2 = sns.boxplot(x ="Algorithm", y="score", data=dataset, palette="viridis", ax=axs[0, 0])
ax_comb2.set_title("Box Plot", size=16)

ax_comb2 = sns.violinplot(x ="Algorithm", y="score", data=dataset, palette="viridis", ax=axs[0, 1])
ax_comb2.set_title("Violin Plot", size=16)

ax_comb2 = sns.histplot(data=dataset, x="score", hue="Algorithm", palette="viridis", multiple="stack", ax=axs[0, 2])
ax_comb2.set_title("Histogram Plot", size=16)

ax_comb2 = sns.kdeplot(data=dataset,  x="score", hue="Algorithm", multiple="fill", palette="viridis",ax=axs[0, 3])
ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)

ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="Algorithm", fill=True, common_norm=False, palette="viridis",alpha=.5,ax=axs[1, 0])
ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)

ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="Algorithm", common_norm=False, palette="viridis", linewidth=3, ax=axs[1, 1])
ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)

ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="Algorithm", cumulative=True, common_norm=False, common_grid=True,palette="viridis", linewidth=3,ax=axs[1, 2])
ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)

ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="Algorithm", multiple="stack", palette="viridis",ax=axs[1, 3])
ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)
```
