```python
plt.figure(figsize=(12,8))
plt.title("KNN 1-15", fontsize= 15)
plt.style.use('seaborn-whitegrid')


# ----- Accuracy -------
ax = sns.lineplot(x=dataset.K,   
             y=dataset.accuracy,       
             data = dataset , 
             color = "#6a0dad", 
             linewidth = 2
             )

# -------- F1 ----------
ax = sns.lineplot(x=dataset.K,   
             y=dataset.F1 ,            
             data = dataset , 
             color = "#0B4619", 
             linewidth = 2
             )

# ------- Recall --------
ax = sns.lineplot(x=dataset.K,   y=dataset.recall,         
             data = dataset , 
             color = "#2C2891", 
             linewidth = 2
             )

# ----- Precision -------
ax = sns.lineplot(x=dataset.K,  
             y=dataset.precision,      
             data = dataset ,
             color = "#C85C5C", 
             linewidth = 2
             )

# ----- Runtime -------
#ax = sns.lineplot(x=dataset.K,  
#             y=dataset.Runtime/240,      
#             data = dataset ,
#             color = "#6a0dad", 
#             linewidth = 2.5
#             )

plt.xlabel("K-value")
plt.ylabel("Score")
ax.legend (loc="upper right")
plt.legend(labels=["Accuracy", "F1", "Recall", "Precision"])
```
