```python
def UNSW_distplot(df, list_of_columns, rows, cols, title, fig_x, fig_y):
    # Categorical Data
    rows = 3  # number of rows
    cols = 4  # number of columns
    c = 1  # initialize plot counter
    sns.set_style("whitegrid")
    print("Creating Distribution plot..")

    fig = plt.figure(figsize=(fig_x, fig_y))
    fig.suptitle(title, size=16)

    for i in col_names:
        plt.subplot(rows, cols, c)
        #plt.title('{}, subplot: {},{},{}'.format(i, a, b, c))
        plt.xlabel(i)
        ax = sns.distplot(df[i])
        c = c + 1

    plt.tight_layout()
    plt.show()
    fig.savefig(f'Analysis_ANN/{title}.png', bbox_inches='tight', dpi=400)
```
