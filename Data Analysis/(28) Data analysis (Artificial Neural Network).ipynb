```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("..")

from Functions.UNSW_DF import * 
```

```python
def read_datasets():
    df1 = pd.read_csv("../ANN/ANN_200E_3L(50-25-1)_AF(Relu-Sigmoid).csv")
    df2 = pd.read_csv("../ANN/ANN_200E_4L(53-36-21-1)_AF(Relu-Sigmoid).csv")
    df3 = pd.read_csv("../ANN/ANN_200E_4L(500-250-100-1)_AF(Relu-Sigmoid).csv")
    df4 = pd.read_csv("../ANN/ANN_200E_5L(100-75-50-25-1)_AF(Relu-Sigmoid).csv")
    df5 = pd.read_csv("../ANN/ANN_200E_7L(200-150-125-100-50-25-1)_AF(Relu-Sigmoid).csv")
    df6 = pd.read_csv("../ANN/ANN_200E_9L(200-175-150-125-100-75-50-25-1)_AF(Relu-Sigmoid).csv")
    df7 = pd.read_csv("../ANN/ANN_200E_11L(300-250-200-175-150-125-100-75-50-25-1)_AF(Relu-Sigmoid).csv")
    return df1, df2, df3, df4, df5, df6, df7
```

```python
# For Barplots, individual Barplot
df1, df2, df3, df4, df5, df6, df7 = read_datasets()

col_drop_dfs = ["no_layers", "runtime(s)", "error_train", "error_test", "precision_train", "F1_train", "recall_train"]
df1 = df1.drop(col_drop_dfs, axis=1)
df2 = df2.drop(col_drop_dfs, axis=1)
df3 = df3.drop(col_drop_dfs, axis=1)
df4 = df4.drop(col_drop_dfs, axis=1)
df5 = df5.drop(col_drop_dfs, axis=1)
df6 = df6.drop(col_drop_dfs, axis=1)
df7 = df7.drop(col_drop_dfs, axis=1)

# Melting datasets
x_label = "epochs"
df1_melt = pd.melt(df1, id_vars=x_label, var_name="metric", value_name="score")
df2_melt = pd.melt(df2, id_vars=x_label, var_name="metric", value_name="score")
df3_melt = pd.melt(df3, id_vars=x_label, var_name="metric", value_name="score")
df4_melt = pd.melt(df4, id_vars=x_label, var_name="metric", value_name="score")
df5_melt = pd.melt(df5, id_vars=x_label, var_name="metric", value_name="score")
df6_melt = pd.melt(df6, id_vars=x_label, var_name="metric", value_name="score")
df7_melt = pd.melt(df7, id_vars=x_label, var_name="metric", value_name="score")
```

```python
# For distplots, combined
df1_m, df2_m, df3_m, df4_m, df5_m, df6_m, df7_m = read_datasets()

col_drop_comb = ["epochs", "runtime(s)", "error_train", "error_test", "precision_train", "F1_train", "recall_train"]
df1_m = df1_m.drop(col_drop_comb, axis=1)
df2_m = df2_m.drop(col_drop_comb, axis=1)
df3_m = df3_m.drop(col_drop_comb, axis=1)
df4_m = df4_m.drop(col_drop_comb, axis=1)
df5_m = df5_m.drop(col_drop_comb, axis=1)
df6_m = df6_m.drop(col_drop_comb, axis=1)
df7_m = df7_m.drop(col_drop_comb, axis=1)

x_label_m = "no_layers"
combined_dataset = pd.concat([df1_m, df2_m, df3_m, df4_m, df5_m, df6_m, df7_m])
combined_dataset_melt = pd.melt(combined_dataset, id_vars=x_label_m, var_name="metric", value_name="score")
```

```python
def create_distplot(dataframe, plot_title="", xsize=30, ysize=24):
    ax_comb2 = sns.set(rc={"figure.figsize":(xsize, ysize), 'axes.facecolor':'white',})
    ax_comb2 = sns.set_style("whitegrid")

    dataset = dataframe
    fig, axs = plt.subplots(ncols=4, nrows=2)
    
    ax_comb2 = sns.boxplot(x ="epochs", y="score", data=dataset, palette="viridis", ax=axs[0, 0])
    ax_comb2.set_title("Box Plot", size=16)

    
    ax_comb2 = sns.violinplot(x ="epochs", y="score", data=dataset, palette="viridis", ax=axs[0, 1])
    ax_comb2.set_title("Violin Plot", size=16)


    ax_comb2 = sns.histplot(data=dataset, x="score", hue="epochs", palette="viridis", multiple="stack", ax=axs[0, 2])
    ax_comb2.set_title("Histogram Plot", size=16)


    ax_comb2 = sns.kdeplot(data=dataset,  x="score", hue="epochs", multiple="fill", palette="viridis",ax=axs[0, 3])
    ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)


    ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="epochs", fill=True, common_norm=False, palette="viridis",alpha=.5,ax=axs[1, 0])
    ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)


    ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="epochs", common_norm=False, palette="viridis", linewidth=3, ax=axs[1, 1])
    ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)


    ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="epochs", cumulative=True, common_norm=False, common_grid=True,palette="viridis", linewidth=3,ax=axs[1, 2])
    ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)


    ax_comb2 = sns.kdeplot(data=dataset, x="score", hue="epochs", multiple="stack", palette="viridis",ax=axs[1, 3])
    ax_comb2.set_title("Kernel density estimate (KDE) Plot", size=16)


    plt.suptitle(plot_title, fontsize=20)
    plt.tight_layout()
    ax_comb2.figure.savefig(f"Analysis_ANN/{plot_title}_distplot.png", bbox_inches='tight', dpi=400)
```
