```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("..")

from Functions.UNSW_DF import * 
```

```python
train, test = DF_original_traintest()
```

```python
# Print the number of train / test samples
print(f"Train data length: {len(train)}")
print(f"Test data length: {len(test)}")
```

```python
ax1 = sns.set(rc={"figure.figsize":(12, 10), 'axes.facecolor':'white',})
ax1 = sns.set_style("whitegrid")

# Create the plots
#fig, axs = plt.subplots(ncols=1, nrows=2)

ax1 = sns.countplot(x="label", data=train, palette="viridis")


# Change the xtick labels for attack / normal
ax1.set_xticklabels(["Normal", "Attack"])
```

```python
train['dataset'] = 'train'
train_label = train[["dataset", "label"]]

test['dataset'] = 'test'
test_label = test[["dataset", "label"]]
```

```python
m_label = pd.concat([train_label, test_label])
```

```python
m_label
```

```python
ax2 = sns.set(rc={"figure.figsize":(12, 7), 'axes.facecolor':'white',})
ax2 = sns.set_style("whitegrid")
ax2 = sns.countplot(x="label", hue="dataset", data=m_label, palette="viridis")

for container in ax2.containers:
    ax2.bar_label(container)


ax2.set_xticklabels(["Normal", "Attack"])
ax2.figure.savefig("Analysis_Dataset/Attack_normal_dist.png", bbox_inches='tight', dpi=400)
```
