```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append("..")

from Functions.UNSW_DF import *
X, Y, x_test, y_test = DF_XY()
train, test = DF_preprocessed_traintest()
```

```python
no_rows = 1000
X, Y, df = X[: no_rows], Y[: no_rows], train[: no_rows]
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
def z_function(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = df["smean"]
y = df["label"]

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)
```

```python
fig = plt.figure(figsize=(20, 20))
ax = plt.axes(projection="3d")
ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
